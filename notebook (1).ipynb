```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
movies = netflix_df[netflix_df['type'] == 'Movie']
print(movies)
movies_90s = movies[movies['release_year'].between(1990,1999)]
print(movies_90s)
```

```python
import matplotlib.pyplot as plt

plt.hist(movies_90s['duration'])
plt.title('Duration of movies in the 90s')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Movies')
plt.show()
```

```python
import numpy as np

# Creating a df with action movies with less than 90 min
actionMovies = movies_90s[movies_90s['genre']== 'Action']
#starting the count
short_movie_count = 0

for label, row in actionMovies.iterrows():
    if row['duration'] < 90:
        short_movie_count =  short_movie_count + 1
    else:
        short_movie_count = short_movie_count
    print(short_movie_count)
```
